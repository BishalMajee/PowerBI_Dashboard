```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\Bishal Majee\Downloads\cleaned.csv")
df.head()
```

```python
# Check for missing values
print(df.isnull().sum())

# Check data types
print(df.dtypes)
```

**Remove duplicate rows

```python
df.drop_duplicates(inplace=True)
df.head()
```

```python
# Distribution of ratings
plt.figure(figsize=(6, 4))
df['rating'].hist(bins=10, edgecolor='black')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

Conclusions:Most Products were rated 5


```python
# Calculate average rating by category
avg_rating_by_category = df.groupby('category')['rating'].mean().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(8, 3))
avg_rating_by_category.plot(kind='bar', color='skyblue')
plt.title('Average Rating by Category')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=65)
plt.show()
```

Conclusions:Accesories & Supplies and Television & Video were the categories with the highest ratings

```python
# Average rating by gender
avg_rating_by_gender = df.groupby('gender')['rating'].mean()

# Plotting
plt.figure(figsize=(6, 4))
avg_rating_by_gender.plot(kind='bar', color='salmon')
plt.title('Average Rating by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Rating')
plt.xticks(rotation=0)
plt.show()


```

```python
df.groupby('year')['amount'].count().plot(kind='bar',title='YearWise Sales')
```

conclusion:From the graph we just plotted we can see that year 2015 had the best sales out of all
years.

```python
# what was the best month of sales
# Filter the dataset to include only records from 2015 to 2018
dataset_2015_2018 = df[(df['year'] >= 2015) & (df['year'] <= 2018)]

# Group the filtered dataset by 'month' and count the number of ratings for each month
monthly_rating_counts = dataset_2015_2018.groupby('month')['rating'].count()

# Create a bar plot to visualize the number of ratings by month
monthly_rating_counts.plot(kind='bar', title='Monthly Rating Counts (2015 to 2018)')

# Show the plot
plt.xlabel('Month')
plt.ylabel('Number of Ratings')
plt.xticks(rotation=0)  # Keep x labels horizontal
plt.show()
#
```

Conclusions:January was the month when most sales were made across the product categories and over the years.

```python
# what brand sold the most in 2015 to 2018
import pandas as pd
import matplotlib.pyplot as plt
# Filter the dataset to include only records from 2015 to 2018
dataset_2015_2018 = df[(df['year'] >= 2015) & (df['year'] <= 2018)]

# Group the filtered dataset by 'brand' and sum the 'amount' for each brand
brand_sales = dataset_2015_2018.groupby('brand')['amount'].sum()

# Sort the total sales amounts in descending order and select the top 10 brands with the highest sales
top_sold_brands = brand_sales.sort_values(ascending=False).head(10)

# Create a bar plot of the top 10 brands by sales
top_sold_brands.plot(kind='bar', title='Brand Wise Top 10 Sales (2015 to 2018)')

# Show the plot
plt.xlabel('Brand')
plt.ylabel('Total Amount Sold')
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.show()
```

Conclusions:Bose was the brand with the most sales in 2015 to 2018 followed by Logitech.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Filter the dataset to include only records from 2015 to 2018
dataset2015_2018 = df[(df['year'] >= 2015) & (df['year'] <= 2018)]

# Group the filtered dataset by 'category' and sum the 'amount' for each category
category_sales = dataset2015_2018.groupby('category')['amount'].sum()

# Sort the total sales amounts in descending order and select the top 10 categories with the highest sales
top_sold_categories = category_sales.sort_values(ascending=False).head(10)

# Create a bar plot of the top 10 most sold product categories
top_sold_categories.plot(kind='bar', title='Top 10 Most Sold Product Categories (2015 to 2018)')

# Show the plot
plt.xlabel('Category')
plt.ylabel('Total Amount Sold')
plt.xticks(rotation=90) 
plt.show()
```

Conclusions:We can see that the category of Headphones sold the most, computers and
accessories were sold the second most while cameras sold the third most

```python
import pandas as pd
import matplotlib.pyplot as plt


# Filter the dataset to include only records from 2015 to 2018
dataset2015_2018 =df[(df['year'] >= 2015) & (df['year'] <= 2018)]

# Group the filtered dataset by 'category' and sum the 'amount' for each category
category_sales = dataset2015_2018.groupby('category')['amount'].sum()

# Sort the total sales amounts in ascending order and select the top 10 categories with the least sales
least_sold_categories = category_sales.sort_values(ascending=True).head(10)

# Create a bar plot of the 10 least sold product categories
least_sold_categories.plot(kind='bar', title='10 Least Sold Product Categories (2015 to 2018)')

# Show the plot
plt.xlabel('Category')
plt.ylabel('Total Amount Sold')
plt.xticks(rotation=90) 
plt.show()
```

Conclusions:We can see that the category of Wearable Technology sold the least followed closely by
Security and Surveillance

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter the dataset for the years 2015 to 2018
dataset2015_2018 = df[(df['year'] >= 2015) & (df['year'] <= 2018)]

# Group by brand, sum the sales amount, and sort in ascending order
least_sold_brands = dataset2015_2018.groupby('brand')['amount'].sum().sort_values(ascending=True)

# Plot the 10 least sold brands
least_sold_brands.head(10).plot(kind='bar', title='10 Least Sold Product Brands (2015 to 2018)')
plt.xlabel('Brand')
plt.ylabel('Total Amount Sold')
plt.show()
```

Conclusions:Archos sold the least followed closely with EINCAR

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by category and sum the sales amount
category_sales = df.groupby('category')['amount'].sum().sort_values(ascending=False)

# Select the top 5 categories
top_5_categories = category_sales.head(5)

# Plot the pie chart
top_5_categories.plot(kind='pie', autopct='%1.1f%%', title='Top 5 Category Sales Percentage', ylabel='')
plt.show()
```

Conclusions:Headphones sales % is the highest followed by Computers & Accessories.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by brand and count the number of ratings (or sales)
brand_sales = df.groupby('brand')['rating'].count().sort_values(ascending=False)

# Select the top 5 brands
top_5_brands = brand_sales.head(5)

# Plot the pie chart
top_5_brands.plot(
    kind='pie',
    autopct='%1.1f%%',
    title='Top 5 Brand Wise Sales Percentage',
    startangle=90,
    colors=['skyblue', 'lightcoral', 'lightgreen', 'lightsalmon', 'lightpink']
)

plt.ylabel('')  # Hide the y-label for better aesthetics
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

Conclusions:Bose and Logitech sales % is the highest followed by Sony

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate gender distribution
gender_distribution = df['gender'].value_counts()

# Plot the pie chart
plt.pie(
    gender_distribution,
    labels=gender_distribution.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['skyblue', 'lightcoral']
)
plt.title('Gender Wise Customer Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Conclusions:Most of the customers are in Female categories
